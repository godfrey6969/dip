```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
path = "/home/sdl16/bike.png"
image = cv2.imread(path)
```

```python
plt.imshow(image)
```

```python
zoom_factor = 40
(r, c, v) = image.shape
zm = np.zeros((r*zoom_factor, c*zoom_factor, v), dtype='uint8')

zm.shape
image.shape
```

```python
for i in range(r*zoom_factor):
    for j in range(c*zoom_factor):
        zm[i][j] = image[math.floor(i / zoom_factor)][math.floor(j / zoom_factor)]

```

```python
plt.imshow(zm)
```

```python
cv2.imwrite('./zoomed_1.png', zm)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

fig, axes = plt.subplots(3, 3, figsize=(10, 10))

for k in range(0, 8):
    factor = pow(2, k)
    f = 256 / factor
    new_img = (gray_image / 256  * factor)
    new_img = np.array(new_img, dtype='uint8')
    ax = axes[(8 - k) // 3, (8 - k) % 3]
    ax.imshow(new_img, cmap="grey")
    ax.axis('off')
plt.show()
```
