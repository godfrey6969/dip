```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "./img1.jpg"
image = cv2.imread(path)
```

```python
gr = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gr.shape
```

```python
neg = np.zeros(shape=gr.shape)
for i in range(gr.shape[0]):
    for j in range(gr.shape[1]):
        neg[i][j] = 255 - gr[i][j]
neg
```

```python
cv2.imwrite("./neg.jpg", neg)
plt.imshow(neg, cmap="gray")
```

```python
lgt = np.zeros(shape=gr.shape)
c = 1
lgt = c * np.log(1 + gr)
lgt = np.array(lgt, dtype="uint8")
lgt
```

```python
plt.imshow(lgt, cmap="gray")
```

```python
gamma = 0.4
c = 255 ** (1 - gamma)
pl = np.zeros(shape=gr.shape)
pl = c * pow(gr, gamma)
pl
```

```python
plt.imshow(pl, cmap="gray")
```

```python
a = 0
b = 255
c = np.min(gr)
d = np.max(gr)
fact = (b - a) / (d - c)
cs = np.zeros(shape=gr.shape)
cs = a + (gr - c) * fact 
cs
```

```python
plt.imshow(cs, cmap="gray")
```

```python
th = 60
max_val = 255
seg = np.zeros(shape=gr.shape)
for i in range(gr.shape[0]):
    for j in range(gr.shape[1]):
        if gr[i][j] > th :
            seg[i][j] = max_val


seg
```

```python
plt.imshow(seg, cmap="gray")
```
